```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
tst = pd.read_csv('activities_cleaned.csv', header = 0)
```

```python
plt.figure(figsize = (10,6)) 
plt.scatter(tst['Moving Duration (Raw Seconds)']/3600, tst['Distance (Raw)'])
plt.xlabel('moving time (hours)') 
plt.ylabel('miles traveled')
#plt.xlim(0,12)
```

```python
run_index = np.where(tst['Activity Type'] == 'Running')
bike_index = np.where(tst['Activity Type'] == 'Road Cycling')
```

```python
plt.figure(figsize = (10,6)) 
plt.scatter(tst['Moving Duration (Raw Seconds)'][run_index[0]]/3600, tst['Distance (Raw)'][run_index[0]], color = 'blue', 
           label= 'runs')
plt.scatter(tst['Moving Duration (Raw Seconds)'][bike_index[0]]/3600, tst['Distance (Raw)'][bike_index[0]],marker = 'x', color = 'red',
            label='bike rides')
plt.xlabel('moving time (hours)') 
plt.ylabel('miles traveled')
plt.legend() 
```
